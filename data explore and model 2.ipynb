```python
# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
# Load model with 8-bit precision (needs less memory)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_8bit=True,  # For lower RAM usage (needs `bitsandbytes`)
    torch_dtype=torch.float16,
)
```

```python
def answer(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    # output = model.generate(**inputs, max_new_tokens=200)
    outputs = model.generate(
        **inputs,
        max_new_tokens=150,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        top_k=50,
        eos_token_id=tokenizer.eos_token_id)
    print(tokenizer.decode(outputs[0], skip_special_tokens=True))
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

# prompting

```python
def extract_instruction_and_answer(text_entry):
    content = text_entry['text']
    start = content.find('[INST]') + len('[INST]')
    end = content.find('[/INST]')
    instruction = content[start:end].strip()
    answer = content[end + len('[/INST]'):].replace('</s>', '').strip()
    return instruction, answer
```

```python
def create_few_shot_prompt_from_dataset(dataset, new_question, new_context, n=30):
    prompt = "<<SYS>>\nYou are a helpful, respectful and honest assistant.\n<</SYS>>\n\n"
    for entry in dataset[:n]:
        instruction, answer = extract_instruction_and_answer(entry)
        prompt += f"[INST] {instruction} [/INST] {answer}</s>\n"

    new_instruction = f"Question: {new_question} Context: {new_context}"
    prompt += f"[INST] {new_instruction} [/INST]"
    return "<s>" + prompt
```

```python
few_shot_prompt = create_few_shot_prompt_from_dataset(
    data,
    new_question="Can wearable devices effectively monitor heart rate variability in post-operative patients?",
    new_context="Heart rate variability (HRV) is an indicator of autonomic nervous system activity. Wearable devices like smartwatches and chest straps have been increasingly used for HRV monitoring. This study evaluates the reliability of wearables for continuous monitoring in post-operative care.",
    n=30  #
)

response = answer(few_shot_prompt)
```

```python
from sentence_transformers import SentenceTransformer, util

# Load a model for semantic similarity
similarity_model = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight and fast
```

```python
import random

def evaluate_model_on_examples(dataset, n=3):
    examples = random.sample(dataset, n)
    results = []

    for entry in examples:
        # Extract question, context, ground-truth
        instruction, true_answer = extract_instruction_and_answer(entry)

        # Create prompt with few-shot examples (excluding this test one)
        few_shot_examples = [ex for ex in dataset if ex != entry][:3]
        question_part = instruction.split("Context:")[0].strip().replace("Question:", "").strip()
        context_part = instruction.split("Context:")[1].strip()

        prompt = create_few_shot_prompt_from_dataset(few_shot_examples, question_part, context_part)
        model_answer = answer(prompt).strip().lower()

        # Compute semantic similarity
        embeddings = similarity_model.encode([true_answer, model_answer], convert_to_tensor=True)
        score = util.cos_sim(embeddings[0], embeddings[1]).item()

        results.append({
            "question": question_part,
            "context": context_part,
            "true_answer": true_answer,
            "model_answer": model_answer,
            "similarity_score": round(score, 4)
        })

    return results
```

```python
results = evaluate_model_on_examples(data, n=4)

# Print results
for r in results:
    print(f"\n🔹 Question: {r['question']}")
    print(f"📘 True Answer: {r['true_answer']}")
    print(f"🤖 Model Answer: {r['model_answer']}")
    print(f"📏 Similarity Score: {r['similarity_score']}")
```
