```python
# Load model with 8-bit precision (needs less memory)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_8bit=True,  # For lower RAM usage (needs `bitsandbytes`)
    torch_dtype=torch.float16,
)
```

```python
def answer(prompt):
    # Tokenize the input prompt using the pre-trained tokenizer and convert to PyTorch tensors
    # Move the tensor to CUDA (GPU) for faster processing
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    # Generate text using the pre-trained language model with the following parameters:
    # - **inputs: the tokenized prompt
    # - max_new_tokens=150: limit generation to 150 new tokens
    # - do_sample=True: enable sampling instead of greedy decoding
    # - temperature=0.7: controls randomness (lower = more deterministic)
    # - top_p=0.9: nucleus sampling parameter (0.9 means considering top 90% probability mass)
    # - top_k=50: consider top 50 most likely tokens at each step
    # - eos_token_id=tokenizer.eos_token_id: stop at end-of-sequence token
    outputs = model.generate(
        **inputs,
        max_new_tokens=150,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        top_k=50,
        eos_token_id=tokenizer.eos_token_id)

    # Print the decoded output (for debugging/logging purposes)
    print(tokenizer.decode(outputs[0], skip_special_tokens=True))

    # Return the decoded output as a string, skipping any special tokens
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
def format_prompt(user_prompt):
    return f"""<s>[INST] <<SYS>>
You are a helpful, respectful and honest assistant.
<</SYS>>

{user_prompt.strip()} [/INST]"""

# Ask the user for input
user_input = input("You: ")
prompt = format_prompt(user_input)

# Tokenize and send to model
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

# Generate response
outputs = model.generate(
    **inputs,
    max_new_tokens=150,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    top_k=50,
    eos_token_id=tokenizer.eos_token_id
)

# Decode and extract assistant's response
decoded = tokenizer.decode(outputs[0], skip_special_tokens=False)

# Remove the prompt part and get only the response
if "[/INST]" in decoded:
    response = decoded.split("[/INST]", 1)[-1].strip()
else:
    response = decoded.replace(prompt, "").strip()

# Show the result
print(f"\n🦙 LLaMA: {response}")
```

```python
def format_for_finetuning(df):
    # Define the output file path for the JSON Lines formatted data
    output_file_path = 'finetune_data.jsonl'

    # Open the output file in write mode with UTF-8 encoding
    with open(output_file_path, 'w', encoding='utf-8') as jsonl_file:
        # Iterate through each row in the DataFrame
        for _, row in df.iterrows():
            # Combine the question with the first context from the contexts list
            # Format: "Question: [question text] Context: [context text]"
            full_input = f"Question: {row['question']} Context: {row['context']['contexts'][0]}"  # Assuming 'contexts' is a list

            # Prepare the data in the specific format required for fine-tuning:
            # 1. <s> - Start of sequence token
            # 2. [INST] - Instruction tag
            # 3. <<SYS>> - System prompt defining the assistant's behavior
            # 4. The combined question and context
            # 5. [/INST] - End of instruction
            # 6. The final decision (model's expected output)
            # 7. </s> - End of sequence token
            formatted_data = {
                "text": f"<s>[INST] <<SYS>>\nYou are a helpful, respectful medical assistant.\n<</SYS>>\n\n{full_input.strip()} [/INST] {row['final_decision'].strip()}</s>"
            }

            # Write each formatted example as a JSON string on a new line in the file
            jsonl_file.write(json.dumps(formatted_data) + '\n')

# Call the function with the training DataFrame
format_for_finetuning(train_df)
```

# fine tuning lora, lora+


```python
import torch
import torch.nn as nn

class LoRALayer(nn.Module):
    def __init__(self, model_dim, rank):
        super(LoRALayer, self).__init__()
        self.rank = rank          # The rank for low-rank decomposition
        self.model_dim = model_dim # Dimension of the model's hidden states

        # Low-rank matrices A and B for adaptation:
        # A projects down to rank dimension, B projects back up
        self.A = nn.Parameter(torch.randn(model_dim, rank))  # Shape: (model_dim × rank)
        self.B = nn.Parameter(torch.randn(rank, model_dim))   # Shape: (rank × model_dim)

    def forward(self, x):
        # Compute low-rank adaptation:
        # 1. x @ A:   (batch_size × seq_len × model_dim) @ (model_dim × rank)
        #    -> (batch_size × seq_len × rank)
        # 2. Then @ B: (batch_size × seq_len × rank) @ (rank × model_dim)
        #    -> (batch_size × seq_len × model_dim)
        low_rank_adaptation = x @ self.A @ self.B

        # Return original input plus adaptation (residual connection)
        return x + low_rank_adaptation
```

```python
# Import the base model class from Hugging Face's transformers library
from transformers import AutoModel

def add_lora_to_model(model, rank):
    """
    Adds LoRA (Low-Rank Adaptation) layers to a transformer model's attention modules.

    LoRA works by adding low-rank matrices to existing weight matrices, enabling efficient
    fine-tuning with fewer parameters than full fine-tuning.

    Args:
        model: The pretrained transformer model to modify
        rank (int): The rank of the low-rank matrices used in LoRA
                   (controls the size and capacity of the adaptation)
    """

    # Iterate through all modules in the model using named_modules()
    # This gives us both the name and the module object for each component
    for name, module in model.named_modules():
        # Check if the current module is a MultiheadAttention layer
        # This is where we want to apply LoRA in most transformer architectures
        if isinstance(module, nn.MultiheadAttention):
            # Get the model dimension (d_model) from the attention module
            model_dim = module.embed_dim

            # Create a new LoRALayer with the appropriate dimensions
            # Note: The actual LoRALayer implementation would need to be defined elsewhere
            lora_layer = LoRALayer(model_dim, rank)

            # Replace the existing self-attention mechanism with our LoRA-modified version
            # This assumes the attention module has a 'self_attention' attribute
            # Note: In practice, the exact replacement might need to be more nuanced
            # depending on the specific model architecture
            setattr(module, 'self_attention', lora_layer)

# Load a pretrained LLaMA-2 7B model from Hugging Face's model hub
# Note: This requires proper authentication and access to the LLaMA-2 models
model = AutoModel.from_pretrained("meta-llama/Llama-2-7b-hf")

# Apply LoRA to the model with a rank of 32
# This rank value is a common starting point, but can be tuned based on specific needs
add_lora_to_model(model, rank=32)
```

```python
def create_few_shot_prompt(data, test_index=10, num_examples=3):
    examples = data[:num_examples]
    test_item = data[test_index]

    prompt = "<<SYS>>\nYou are a helpful biomedical QA assistant.\n<</SYS>>\n\n"

    for ex in examples:
        ex_split = ex['text'].split('[/INST]')
        prompt += f"[INST]{ex_split[0].split('[INST]')[1].strip()}[/INST] {ex_split[1].strip()}\n\n"

    test_prompt = test_item['text'].split('[/INST]')[0].split('[INST]')[1].strip()
    prompt += f"[INST] {test_prompt} [/INST]"

    return prompt, test_item
```
